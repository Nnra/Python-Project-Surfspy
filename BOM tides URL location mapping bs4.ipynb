```python
import urllib
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
main_url = 'http://www.bom.gov.au/australia/tides/'
```

```python
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(main_url, headers=hdr)

page = urlopen(req)
soup_tides_main = BeautifulSoup(page, "lxml")
fout = open('tides_main.txt', 'wt', encoding='utf-8')
fout.write(str(soup_tides_main)) # save to file
fout.close()
```

```python
soup_tides_main
```

```python
soup_tides_main.find_all("a", class_="feature")
```

```python
location_id = [x.get('id') for x in soup_tides_main.find_all("a", class_="feature")]
```

```python
location_id
```

```python
location_name = [x.text for x in soup_tides_main.find_all("a", class_="feature")]
```

```python
location_name
```

```python
d = {'location_id':location_id, 'location_name':location_name}
df_location = pd.DataFrame(data=d)
df_location
```

```python
df_location['state'] = pd.Series(location_id).str.split('_', expand=True)
```

```python
df_location
```

```python
df_location['url_name'] = df_location['state'].str.lower() + '-' + pd.Series(df_location['location_name']).str.replace(' ','-').str.lower().str.replace('.','')
```

```python
df_location
```

```python
df_location.to_csv('coast_name.csv', index=False)
```
